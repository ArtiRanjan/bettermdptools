# Mountain Car
The gym Mountain Car environment has a continuous state space and does not include P, the transition rewards matrix. We'll use the DiscreteMountainCar class to create a wrapped gym environment, which discretizes the observation space and includes P - which is necessary for the bellman equation, used in value iteration and policy iteration (and other algorithms).

Important note: DiscreteMountainCar implements a modified mountain car environment where the acceleration force and max speed are increased by 50%. This allows for much faster convergence. Implementations of VI/PI/Q-learning are also highly vectorized.

It is recommended to copy the following code into a separate python script and run it there, as jupyter notebook executes it much slower. Be sure to experiment with different bin sizes and gamma values.
